```python
# Classe Livro (livro.py)

class Livro:
    def __init__(self, titulo: str, autor: str, ano: int):         #o que é init? O que é def?
        self.__titulo = titulo          # é assim que se monta a biblioteca?
        self.__autor = autor
        self.__ano = ano
        self.__disponivel = True # O Status Inicial

    #Métodos para acessar atributos (getters)
    @property # O que é isso??
    def titulo(self):
        return self.__titulo

    @property
    def disponivel(self):
        return self.__disponivel

    #Métodos para emprestimo/devolução
    def emprestar (self):
        if self.__disponivel:
            self.__disponivel = False
            return True
        return False # Já está emprestado

    def devolver(self):
        self.__disponivel = True
```

```python
#Classe Usuario (usuario.py)

class Usuario:
    def __init__(self, nome: str, email: str): #O self é a identidade para colocar as coisas suas, em nome e email
        self.__nome = nome #Colocar seu nome aqui
        self.__email = email #Colocar seu email aqui
        self.__livros_emprestados = [] # Lista de livros

    @property
    def nome(self):
        return self.__nome

    def emprestar_livro(self, livro):
        if livro.emprestar(): #Só empresta quando estiver disponível?
            self.__livros_emprestados.append(livro)
            return True
        return False # Livro não está com o usuário

    def devolver_livro(self, livro):
        if livro in self.__livros_emprestados:
            livro.devolver()
            self.__livros_emprestados.remove(livro)
            return True
        return False # Livro não estava com o usuário
```

```python
#Classe Biblioteca (biblioteca.py)

from livro import Livro
from usuario import Usuario

class Biblioteca:
    def __init__(self):
        self.__livros = [] #Lista de livros (pode ser substituída por lista encadeada)
        self.__usuarios = [] #lista de usuários

    def adicionar_livro(self, livro: Livro):
        self.__livros.append(livro)

    def buscar_livro(self, termo: str):
        return [livro for livro in self.__livros
                if termo.lower() in livro.titulo.lower() or
                   termo.lower() in livro.autor.lower]

    def relatorio_emprestimos(self):
        for usuario in self.__usuarios:
            for livro in usuario.livros_emprestados:
                print(f"Livro: {livro.titulo} | Usuário: {usuario.nome}")
```

```python
#Livros de Herança em Livro.py

class LivroDigital(Livro):
    def __init__(self, titulo: str, autor: str, ano: int, formato: str): #Definição de si com string de titulo, ano e formado
        super().__init__(titulo, autor, ano)
        self.__formato = formato # Ex: PDF, EPUB

    def emprestar(self):
        print("Livro digital não precisa de devolução física!")
        return super().emprestar()
```

```python
#Codigo copiado.

class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None

class ListaEncadeada:
    def __init__(self):
        self.cabeca = None

    def adicionar(self, valor):
        novo_no = No(valor)
        if not self.cabeca:
            self.cabeca = novo_no
        else:
            atual = self.cabeca
            while atual.proximo:
                atual = atual.proximo
            atual.proximo = novo_no
```
